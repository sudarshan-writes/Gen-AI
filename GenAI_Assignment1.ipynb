**TASK 1: Replace Hardcoded Prompt**

```python

# Step 1: Import PromptTemplate (correct import to avoid error)
from langchain_core.prompts import PromptTemplate


# Step 2: Create Reusable PromptTemplate
# This replaces hardcoded prompt with dynamic template

task1_template = PromptTemplate(
    input_variables=["topic"],   # dynamic input
    template="Explain {topic} in simple terms for beginners"
)

# Step 3: Create Function
# This function uses template to generate prompt dynamically

def explain_topic(topic):

    # Generate prompt using template
    final_prompt = task1_template.format(topic=topic)

    return final_prompt
    
# Step 4: Test Cases

print("Task 1 Output:\n")

print(explain_topic("Artificial Intelligence"))
print(explain_topic("Neural Network"))
print(explain_topic("Machine Learning"))

print("\n")
```

**TASK 2: Multi-Input Prompt System**

```python

# Import PromptTemplate (CORRECT import for latest LangChain)
from langchain_core.prompts import PromptTemplate


# Step 1: Create a reusable template

# input_variables defines dynamic inputs (MANDATORY as per assignment)
multi_input_template = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Explain {topic} for {audience} in a {tone} tone"
)

# Step 2: Create a function (modular design)

# This function generates prompt dynamically
def generate_multi_input_prompt(topic, audience, tone):

    # Using .format() (NO f-strings as per rules)
    prompt = multi_input_template.format(
        topic=topic,
        audience=audience,
        tone=tone
    )

    return prompt


# Step 3: Test Cases (as given in assignment)

print("Task 2 Outputs:\n")

# Test Case 1
output1 = generate_multi_input_prompt("AI", "beginners", "friendly")
print(output1)

# Test Case 2
output2 = generate_multi_input_prompt("Python", "kids", "fun")
print(output2)

# Test Case 3
output3 = generate_multi_input_prompt("Deep Learning", "engineers", "technical")
print(output3)
```

**TASK 3: Prompt Variations Engine**

```python

# Import PromptTemplate (correct import)
from langchain_core.prompts import PromptTemplate

# Step 1: Create 3 different templates


# Teaching Template
teaching_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} clearly step by step"
)

# Interview Template
interview_template = PromptTemplate(
    input_variables=["topic"],
    template="Ask 3 questions about {topic}"
)

# Storytelling Template
story_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} as a story"
)

# Step 2: Create a function (modular & reusable)

def generate_prompt_variations(topic):

    # Generate prompt for Teaching style
    teaching_prompt = teaching_template.format(topic=topic)

    # Generate prompt for Interview style
    interview_prompt = interview_template.format(topic=topic)

    # Generate prompt for Storytelling style
    story_prompt = story_template.format(topic=topic)

    # Store all outputs in dictionary (clean structure)
    return {
        "Teaching": teaching_prompt,
        "Interview": interview_prompt,
        "Storytelling": story_prompt
    }

# Step 3: Run for given topic

print("Task 3 Outputs:\n")

# Given in assignment
topic_input = "Machine Learning"

# Call function
results = generate_prompt_variations(topic_input)

# Print outputs
for key, value in results.items():
    print(key + ":", value)
```

**TASK 4: ChatPromptTemplate System**

```python


# Correct import
from langchain_core.prompts import ChatPromptTemplate

# Step 1: Function to define system behavior (LOGIC SEPARATE)

def get_system_message(role):

    # Role-based behavior (logic layer)
    if role == "teacher":
        return "You are a teacher who explains concepts clearly"

    elif role == "interviewer":
        return "You are an interviewer who asks questions"

    elif role == "motivator":
        return "You are a motivator who inspires people"

    else:
        return "You are a helpful assistant"

# Step 2: Chat Prompt Generator Function

def generate_chat_prompt(role, topic):

    # Get system message dynamically
    system_message = get_system_message(role)

    # Create ChatPromptTemplate
    chat_template = ChatPromptTemplate.from_messages([
        ("system", system_message),        # System defines behavior
        ("human", "{topic}")               # User input (dynamic)
    ])

    # Generate messages
    messages = chat_template.format_messages(topic=topic)

    return messages

# Step 3: Example (as per assignment)

print("Task 4 Output:\n")

role_input = "teacher"
topic_input = "Neural Networks"

output = generate_chat_prompt(role_input, topic_input)

# Clean printing (IMPORTANT for evaluation)
for msg in output:
    print(msg.__class__.__name__ + ":", msg.content)
```

**TASK 5: Input Validation Layer**

```python

def validate_inputs(audience, tone):

    # Valid values
    valid_audience = ["beginner", "intermediate", "expert"]
    valid_tone = ["formal", "casual", "fun"]

    # Default values
    default_audience = "beginner"
    default_tone = "casual"

    # Check audience
    if audience not in valid_audience:
        print("Warning: '{}' is invalid. Defaulting to '{}'.".format(audience, default_audience))
        audience = default_audience

    # Check tone
    if tone not in valid_tone:
        print("Warning: '{}' is invalid. Defaulting to '{}'.".format(tone, default_tone))
        tone = default_tone

    return audience, tone

# Test Cases (As Expected Output)

print("Outputs:\n")

test_cases = [
    ("expert", "angry"),
    ("novice", "fun"),
    ("beginner", "formal"),
    ("master", "serious")
]

for audience, tone in test_cases:
    print("Testing -> Audience: '{}', Tone: '{}'".format(audience, tone))

    final_audience, final_tone = validate_inputs(audience, tone)

    print("Result  -> Audience: {}, Tone: {}\n".format(final_audience, final_tone))
```

**Task 6: Prompt Generator App**

```python

# Import PromptTemplate (UPDATED import to avoid error)
from langchain_core.prompts import PromptTemplate

# Reuse validation from Task 5

def validate_inputs(audience, tone):

    valid_audience = ["beginner", "intermediate", "expert"]
    valid_tone = ["formal", "casual", "fun"]

    default_audience = "beginner"
    default_tone = "casual"

    if audience not in valid_audience:
        print("Warning: '{}' is invalid. Defaulting to '{}'.".format(audience, default_audience))
        audience = default_audience

    if tone not in valid_tone:
        print("Warning: '{}' is invalid. Defaulting to '{}'.".format(tone, default_tone))
        tone = default_tone

    return audience, tone

# Define Prompt Templates

# Teaching Template
teaching_template = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Explain {topic} for {audience} in a {tone} tone step by step"
)

# Interview Template
interview_template = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Ask 3 interview questions about {topic} suitable for {audience} in a {tone} tone"
)

# Storytelling Template
story_template = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Explain {topic} as a story for {audience} in a {tone} tone"
)

# Main Function
def generate_prompt(topic, audience, tone, style):

    # Step 1: Validate input
    audience, tone = validate_inputs(audience, tone)

    # Step 2: Select template based on style
    if style == "teaching":
        selected_template = teaching_template
    elif style == "interview":
        selected_template = interview_template
    elif style == "storytelling":
        selected_template = story_template
    else:
        return "Invalid style. Choose teaching, interview, or storytelling."

    # Step 3: Generate prompt
    final_prompt = selected_template.format(
        topic=topic,
        audience=audience,
        tone=tone
    )

    return final_prompt

# Test Cases

print("Task 6 Output:\n")

print(generate_prompt("Neural Networks", "beginner", "fun", "storytelling"))
print()
print(generate_prompt("Machine Learning", "expert", "formal", "teaching"))
print()
print(generate_prompt("Python", "kids", "casual", "interview"))
```

**Task 7: Template Reusability Test**

```python

# Import PromptTemplate (safe import)
from langchain_core.prompts import PromptTemplate

# ONE Template Only

reusable_template = PromptTemplate(
    input_variables=["topic", "audience", "tone", "style"],
    template="Explain {topic} for {audience} in a {tone} {style} style"
)


# Function for reuse

def generate_reusable_prompt(topic, audience, tone, style):

    return reusable_template.format(
        topic=topic,
        audience=audience,
        tone=tone,
        style=style
    )

# 5 Different Test Inputs

print("Task 7 Output:\n")

print(generate_reusable_prompt("Artificial Intelligence", "beginner", "fun", "teaching"))
print()

print(generate_reusable_prompt("Python", "intermediate", "casual", "interview"))
print()

print(generate_reusable_prompt("Data Science", "expert", "formal", "storytelling"))
print()

print(generate_reusable_prompt("Machine Learning", "beginner", "casual", "teaching"))
print()

print(generate_reusable_prompt("Deep Learning", "expert", "fun", "storytelling"))
```
